```python
from random import random
from math import cos, sin, pi
import matplotlib.pyplot as plt
```

```python
#Construção do circulo
def circulo(x = 0.5, y = 0.5, r = 0.5):
    a = [ x + r*cos(pi*ang/180) for ang in range(361)]
    b = [ y + r*sin(pi*ang/180) for ang in range(361)]
    return a,b
```

```python
#n pontos aleatórios
def pontos(n = 0):
    a = [random() for i in range(n)]
    b = [random() for i in range(n)]
    return a,b
```

```python
#Desenho do circulo
x, y = circulo()

#Pontos aleatórios
k, j = pontos(1000)
```

```python
#Desenhos
plt.plot(x,y)#Circulo
plt.plot(k,j, linestyle = ' ', color = 'r', marker = '.')#Pontos aleatórios

#Design
plt.title("Gráfico")
plt.grid(True)

#Conclusão
plt.legend()
plt.show()
```
