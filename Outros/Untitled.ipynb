```python
import matplotlib.pyplot as plt
import math
```

```python
#Construção do circulo
def circulo(x = 0, y = 0, r = 1):
    a = [ x + r*math.cos(math.pi*ang/180) for ang in range(361)]
    b = [ y + r*math.sin(math.pi*ang/180) for ang in range(361)]
    return a,b
```

```python
x, y = circulo()
```

```python
plt.plot(x,y)

plt.grid(False)

plt.legend()
plt.show()
```
